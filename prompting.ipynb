```python
import google.generativeai as genai
import yaml
with open('config.yaml', 'r') as f:
    config = yaml.safe_load(f)
GEMINI_API_KEY = config['GEMINI_API_KEY']
genai.configure(api_key=GEMINI_API_KEY)
```

```python
def save_file(filename: str, content: str) -> str:
    """
    Saves raw text in the content string to a file specified by filename string
    """
    with open(filename, 'w') as f:
        f.write(content)

def read_file(filename: str) -> str:
    """
    Reads the content of the specified file and returns it as a string.
    """
    try:
        with open(filename, 'r', encoding='utf-8') as f:
            return f.read()
    except FileNotFoundError:
        return f"Error: The file '{filename}' was not found."
    except Exception as e:
        return f"Error: {e}"

model = genai.GenerativeModel('gemini-2.0-flash-exp',tools=[save_file, read_file])
model
```

```python
prompt = """
Create a lightweight, interactive floating paw cursor effect for all the pages. 
The effect should feature small, orange paw prints that follow the user's cursor
in real-time. The paws should trail behind the cursor, with 5-6 paw prints 
appearing in sequence, not too far apart. As the cursor moves, older paw prints 
should smoothly fade out, creating a subtle, non-intrusive animation. The effect 
should work on all screen sizes, be optimized for performance, and feel natural 
with smooth transitions and slight randomness in paw placement to mimic organic 
movement.
"""

chat = model.start_chat(enable_automatic_function_calling=True)
chat.send_message(prompt)
```
